```python
%%time
# a running example to compare efficiency
lst=list(range(1,100000))
arr=np.array(lst)
arr.sum()
```

```python
%%time
#列表中每个数完成乘方，再开根号，再求和
import math
for i in range(100):
    lst_result=sum([math.sqrt(item**2) for item in lst])
print(lst_result)
```
